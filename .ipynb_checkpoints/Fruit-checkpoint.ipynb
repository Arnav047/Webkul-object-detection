```python
import os
import shutial
import random

DATASET_DIR = "dataset"
IMAGES_DIR = os.path.join(DATASET_DIR, "images")
LABELS_DIR = os.path.join(DATASET_DIR, "labels")

IMAGE_EXTS = ['.jpg', '.jpeg', '.png']  # lower case
SPLIT_RATIO = 0.8  # 80% train, 20% test

for split in ['train', 'test']:
    os.makedirs(os.path.join(IMAGES_DIR, split), exist_ok=True)
    os.makedirs(os.path.join(LABELS_DIR, split), exist_ok=True)

image_files = [
    f for f in os.listdir(IMAGES_DIR)
    if os.path.splitext(f)[1].lower() in IMAGE_EXTS
    and os.path.isfile(os.path.join(IMAGES_DIR, f))
]

random.shuffle(image_files)
split_idx = int(len(image_files) * SPLIT_RATIO)
train_images = image_files[:split_idx]
test_images = image_files[split_idx:]

def move_pair(files, split):
    for img in files:
        base = os.path.splitext(img)[0]
        img_path = os.path.join(IMAGES_DIR, img)
        label_path = os.path.join(LABELS_DIR, base + ".txt")

        img_dest = os.path.join(IMAGES_DIR, split, img)
        label_dest = os.path.join(LABELS_DIR, split, base + ".txt")

        shutil.move(img_path, img_dest)

        if os.path.exists(label_path):
            shutil.move(label_path, label_dest)
        else:
            print(f" Warning: No label found for {img}")

move_pair(train_images, 'train')
move_pair(test_images, 'test')

print("Dataset successfully split into train/test sets!")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') 

model.train(
    data='C:/Users/ADMIN/Documents/Custom Object Detector/dataset/data.yaml',
    epochs=50,
    imgsz=416,
    batch=4,
    
    fliplr=0.5,          # Horizontal flip (common & safe)
    flipud=0.0,          # No vertical flip (unnatural for objects like fork/spoon)
    hsv_h=0.015,         # Hue variation
    hsv_s=0.5,           # Saturation (moderate)
    hsv_v=0.4,           # Brightness (moderate)
    translate=0.05,      # Slight translation
    scale=0.2,           # Slight zoom in/out
    shear=0.0,           # No shearing
    perspective=0.0,     # Disable to avoid geometric distortion
    mosaic=0.5,          # Enable mosaic moderately
    mixup=0.0,           # Disable MixUp (can confuse model on small datasets)
    copy_paste=0.0,      # Disable for now
    patience=10,
)
```

```python
%matplotlib inline
from PIL import Image
import matplotlib.pyplot as plt
import os

results_path = 'runs/detect/train/results.png'

if os.path.exists(results_path):
    img = Image.open(results_path)
    plt.figure(figsize=(10, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title("Training Results: Loss & mAP curves")
    plt.show()
else:
    print("results.png not found. Check your path or training output folder.")
```

```python
metrics = model.val()
precision_per_class = metrics.box.p
recall_per_class = metrics.box.r
f1_per_class = metrics.box.f1

mean_precision = sum(precision_per_class) / len(precision_per_class)
mean_recall = sum(recall_per_class) / len(recall_per_class)
mean_f1 = sum(f1_per_class) / len(f1_per_class)

print("Precision per class:", precision_per_class)
print("Recall per class:", recall_per_class)
print("F1 score per class:", f1_per_class)

print("\nMean Precision:", round(mean_precision, 3))
print("Mean Recall:", round(mean_recall, 3))
print("Mean F1 Score:", round(mean_f1, 3))
```

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt

image_pairs = [
    ('dataset/images/test/apple_2.jpg', 'result_apple_2.jpg'),
    ('dataset/images/test/orange_23.jpg', 'result_orange_23.jpg'),
    ('dataset/images/test/Kiwi_6.jpg', 'result_Kiwi_6.jpg')  
]

plt.figure(figsize=(12, 4 * len(image_pairs)))

for i, (input_path, output_path) in enumerate(image_pairs):
    input_img = cv2.imread(input_path)
    input_rgb = cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB) if input_img is not None else None

    output_img = cv2.imread(output_path)
    output_rgb = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB) if output_img is not None else None

    plt.subplot(len(image_pairs), 2, 2 * i + 1)
    if input_rgb is not None:
        plt.imshow(input_rgb)
        plt.title(f"Input: {input_path.split('/')[-1]}")
    else:
        plt.text(0.5, 0.5, "Input Image Not Found", ha='center', va='center')
    plt.axis('off')

    plt.subplot(len(image_pairs), 2, 2 * i + 2)
    if output_rgb is not None:
        plt.imshow(output_rgb)
        plt.title(f"Output: {output_path.split('/')[-1]}")
    else:
        plt.text(0.5, 0.5, "Output Image Not Found", ha='center', va='center')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
